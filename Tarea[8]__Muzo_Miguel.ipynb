```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos proporcionados
x = np.array([0.2, 0.3, 0.6, 0.9, 1.1, 1.3, 1.4, 1.6])
y = np.array([0.050446, 0.098426, 0.33277, 0.72660, 1.0972, 1.5697, 1.8487, 2.5015])

# Función para calcular el error cuadrático medio
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

# Ajuste del polinomio de grado 1
coef_1 = np.polyfit(x, y, 1)
poly_1 = np.poly1d(coef_1)
y_pred_1 = poly_1(x)
error_1 = mse(y, y_pred_1)

# Ajuste del polinomio de grado 2
coef_2 = np.polyfit(x, y, 2)
poly_2 = np.poly1d(coef_2)
y_pred_2 = poly_2(x)
error_2 = mse(y, y_pred_2)

# Ajuste del polinomio de grado 3
coef_3 = np.polyfit(x, y, 3)
poly_3 = np.poly1d(coef_3)
y_pred_3 = poly_3(x)
error_3 = mse(y, y_pred_3)

# Ajuste de la forma be^ax
def exp_func(x, a, b):
    return b * np.exp(a * x)

params_exp, _ = curve_fit(exp_func, x, y, p0=(0.1, 1))
y_pred_exp = exp_func(x, *params_exp)
error_exp = mse(y, y_pred_exp)

# Ajuste de la forma bx^a
def power_func(x, a, b):
    return b * x**a

params_power, _ = curve_fit(power_func, x, y, p0=(1, 1))
y_pred_power = power_func(x, *params_power)
error_power = mse(y, y_pred_power)

# Imprimir los resultados de las ecuaciones y errores
print(f'Polinomio grado 1: y = {coef_1[0]:.5f}x + {coef_1[1]:.5f}')
print(f'Error cuadrático medio: {error_1:.5f}\n')

print(f'Polinomio grado 2: y = {coef_2[0]:.5f}x^2 + {coef_2[1]:.5f}x + {coef_2[2]:.5f}')
print(f'Error cuadrático medio: {error_2:.5f}\n')

print(f'Polinomio grado 3: y = {coef_3[0]:.5f}x^3 + {coef_3[1]:.5f}x^2 + {coef_3[2]:.5f}x + {coef_3[3]:.5f}')
print(f'Error cuadrático medio: {error_3:.5f}\n')

print(f'Exponencial: y = {params_exp[1]:.5f}e^({params_exp[0]:.5f}x)')
print(f'Error cuadrático medio: {error_exp:.5f}\n')

print(f'Potencia: y = {params_power[1]:.5f}x^{params_power[0]:.5f}')
print(f'Error cuadrático medio: {error_power:.5f}\n')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos extraídos de la tabla
weights = np.array([4800, 3700, 3400, 2800, 1900])
percentages = np.array([3.1, 4.0, 5.2, 6.4, 9.6])

# Calcular la recta de regresión por mínimos cuadrados
A = np.vstack([weights, np.ones(len(weights))]).T
m, c = np.linalg.lstsq(A, percentages, rcond=None)[0]

# Generar puntos para la recta de regresión
regression_line = m * weights + c

# Plotting the data points and the regression line
plt.scatter(weights, percentages, color='blue', label='Datos originales')
plt.plot(weights, regression_line, color='red', label='Recta de regresión')
plt.xlabel('Peso promedio (lb)')
plt.ylabel('Porcentaje de presentación (%)')
plt.title('Ajuste de regresión lineal')
plt.legend()
plt.grid(True)
plt.show()

m, c
```
